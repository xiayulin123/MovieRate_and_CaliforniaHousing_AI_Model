# Setting Up the Environment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Get Data from the dataset

```python
data = pd.read_csv("Movies_data.csv")
```

```python
#check the data
data
```

```python
#check data if there is any part missing
data.info()
```

# Seperate the Data into X and Y
## X = independents Y = dependent

```python
from sklearn.model_selection import train_test_split

#define what x and y is
X = data.drop(['vote_average'], axis=1) #x is the table without vote_average
Y = data['vote_average'] #y is the table only with vote_average
```

```python
X
```

```python
Y
```

# Seperate into train dataset and test dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
#combine the data to traindata set
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
train_data.original_language.value_counts()
```

## Randomize language data: Use appropriate encoding (e.g., one-hot, embeddings)...

## Transform release date to a numerical format (e.g., timestamp, days since reference date).

```python
train_data['release_date'] = pd.to_datetime(train_data['release_date'])
train_data['Year'] = train_data['release_date'].dt.year
train_data['Month'] = train_data['release_date'].dt.month
train_data['Day'] = train_data['release_date'].dt.day
```

```python
train_data
```

## Omit the "Release Date" column, as its information has already been separated into individual columns for Year, Month, and Day.

```python
train_data.drop('release_date', axis=1, inplace=True)
```

```python
train_data
```

## As the title is not easily analyzable and could differ from others (as the train_data is randomly selected from 20% of the entire dataset).

```python
train_data['title_Length'] = train_data['title'].apply(len)
```

```python
train_data.drop('title', axis=1, inplace=True)
```

```python
# one-hot encoding
# train_data = pd.get_dummies(train_data, columns=['original_language'], prefix='lang')
train_data['is_english'] = train_data['original_language'].apply(lambda lang: 1 if lang == 'en' else 0)
```

```python
train_data.drop('original_language', axis =1, inplace = True)
```

```python
train_data.corr()
```

## Generate a Heatmap to visualize factor correlations

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x = "popularity", y ="Year", data= train_data, hue = "vote_average", palette="coolwarm")
```

# Model Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train, y_train = train_data.drop(['vote_average'], axis=1), train_data['vote_average']
X_train_s = scaler.fit_transform(X_train)
reg = LinearRegression()

reg.fit(X_train_s, y_train)
```

```python
X_train
```

```python
X_test
```

```python
test_data = X_test.join(y_test)

test_data['release_date'] = pd.to_datetime(test_data['release_date'])
test_data['Year'] = test_data['release_date'].dt.year
test_data['Month'] = test_data['release_date'].dt.month
test_data['Day'] = test_data['release_date'].dt.day
test_data.drop('release_date', axis=1, inplace=True)
test_data['title_Length'] = test_data['title'].apply(len)
test_data.drop('title', axis=1, inplace=True)
test_data['is_english'] = test_data['original_language'].apply(lambda lang: 1 if lang == 'en' else 0)
test_data.drop('original_language', axis=1, inplace=True)
```

```python
X_test, y_test = test_data.drop(['vote_average'], axis=1), test_data['vote_average']
```

```python
X_test
```

```python
X_test_s = scaler.fit_transform(X_test)
```

```python
reg.score(X_test_s, y_test)
```

```python
plt.figure(figsize=(6, 4))
predicted_values = reg.predict(X_test_s)
plt.scatter(y_test, predicted_values, color='blue', alpha=0.5)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Predicted vs. Actual Values')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

# Model Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test_s, y_test)
```

# Model Cross Validation

```python
from sklearn.model_selection import GridSearchCV

fores = RandomForestRegressor()
param_grid = {
    "n_estimators":[3, 10, 30],
    "max_features":[2,4,6,8]
}
grid_search = GridSearchCV(forest, param_grid, cv =5, scoring = "neg_mean_squared_error", return_train_score = True)
grid_search.fit(X_train_s, y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(X_test_s, y_test)
```
